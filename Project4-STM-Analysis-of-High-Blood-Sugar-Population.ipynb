# Statistical analysis of Diabetic population in a specific region

### Histogram showing Agewise Distrbution of High Blood Sugar among the population

```python
# Sample data for each age group and corresponding high blood sugar cases
age_groups = ['0-18', '19-35', '36-50', '51-65', '66+']
high_blood_sugar_cases = [50, 200, 400, 300, 250]

# Mapping age groups to numerical values for histogram plotting
age_bins = [1, 2, 3, 4, 5]  # Representing each age group with a bin number

# Plotting
plt.figure(figsize=(10, 6))
plt.hist(age_bins, bins=len(age_groups), weights=high_blood_sugar_cases, edgecolor='black', color='skyblue', rwidth=0.8)
plt.xlabel('Age Group')
plt.ylabel('Number of High Blood Sugar Cases')
plt.title('Histogram of High Blood Sugar Cases by Age Group')
plt.xticks(ticks=age_bins, labels=age_groups)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

### Histogram showing frequency distribution of High Blood Sugar population by their levels

```python
import numpy as np
from scipy.stats import skew, kurtosis


# Sample data for high blood sugar readings
# Let's assume these readings are collected from the population 
high_blood_sugar_readings = [90, 97, 130, 162, 158, 155, 160, 170, 100, 190, 85, 210, 220, 230, 300, 250, 400, 270, 280, 290]

# Convert the list to a numpy array
data_array = np.array(high_blood_sugar_readings)

# Calculate descriptive statistics using numpy
mean = np.mean(data_array)
median = np.median(data_array)
variance = np.var(data_array, ddof=1)  # ddof=1 provides the sample variance
std_dev = np.std(data_array, ddof=1)  # ddof=1 provides the sample standard deviation
data_range = np.ptp(data_array)  # Peak to Peak (range: max - min)

# Calculate skewness and kurtosis using scipy.stats
data_skewness = skew(data_array)
data_kurtosis = kurtosis(data_array)

# Display the statistics
print("Descriptive Statistics using numpy:")
print(f"Mean: {mean:.2f}")
print(f"Median: {median:.2f}")
print(f"Variance: {variance:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")
print(f"Range: {data_range}")
print(f"Skewness: {data_skewness:.2f}")
print(f"Kurtosis: {data_kurtosis:.2f}")

# Define the central point for normal blood sugar range (e.g., 100 mg/dL)
central_point = 100

# Plotting the histogram
plt.figure(figsize=(12, 6))
plt.hist(high_blood_sugar_readings, bins=5, edgecolor='black', color='skyblue', rwidth=0.8)
plt.axvline(central_point, color='red', linestyle='dashed', linewidth=2, label=f'Normal Range Center ({central_point} mg/dL)')
plt.xlabel('Blood Sugar Levels (mg/dL)')
plt.ylabel('Frequency')
plt.title('Histogram of High Blood Sugar Readings with Central Point')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

# Show skewness
mean_reading = np.mean(high_blood_sugar_readings)
skewness = mean_reading - central_point
plt.axvline(mean_reading, color='green', linestyle='dashed', linewidth=2, label=f'Mean Reading ({mean_reading:.2f} mg/dL)')
plt.legend()

# Display the plot
plt.show()
```
